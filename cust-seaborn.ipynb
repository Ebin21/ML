```python
import pandas as pd 
import seaborn as sns

cust_data=pd.read_csv('customer.csv')
cust_data.head(1)
```

```python
sns.displot(cust_data.Age, color='red', kde=True)
```

```python
cust_data['Salary'] = 1000 + (cust_data.Age * 150)
cust_data['Salary']
```

```python
sns.displot(cust_data.Salary, kde=True, color='green')
```
